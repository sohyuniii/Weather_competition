```python
import pandas as pd
import numpy as np
import os

train = pd.read_csv('./train.csv')
test = pd.read_csv('./test.csv')
cluster_df = pd.read_csv('./cluster_df.csv')
train['train'] = 1
test['train'] = 0

df = pd.concat([train, test])
df = pd.merge(df, cluster_df, how = 'left')
print(df.shape)
print(df.columns)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
plt.figure(figsize=(12,4))
plt.subplot(1,2,1)
sns.distplot(df[df.cluster == 0]["y"], color="blue",hist=False)
sns.distplot(df[df.cluster == 1]["y"], color="orange",hist=False)
sns.distplot(df[df.cluster == 2]["y"], color="green",hist=False)
plt.subplot(1,2,2)
sns.boxplot(df.cluster, df.y)
```

```python
plt.figure(figsize=(12,4))
plt.subplot(1,2,1)
sns.distplot(df[df.cluster == 0]["road_num"], color="blue",hist=False)
sns.distplot(df[df.cluster == 1]["road_num"], color="orange",hist=False)
sns.distplot(df[df.cluster == 2]["road_num"], color="green",hist=False)
plt.subplot(1,2,2)
sns.boxplot(df.cluster, df.road_num)
```

```python
plt.figure(figsize=(12,4))
plt.subplot(1,2,1)
sns.distplot(df[df.cluster == 0]["accident_count"], color="blue",hist=False)
sns.distplot(df[df.cluster == 1]["accident_count"], color="orange",hist=False)
sns.distplot(df[df.cluster == 2]["accident_count"], color="green",hist=False)
plt.subplot(1,2,2)
sns.boxplot(df.cluster, df.accident_count)
```

```python
plt.figure(figsize=(12,4))
plt.subplot(1,2,1)
sns.distplot(df[df.cluster == 0]["severity"], color="blue",hist=False)
sns.distplot(df[df.cluster == 1]["severity"], color="orange",hist=False)
sns.distplot(df[df.cluster == 2]["severity"], color="green",hist=False)
plt.subplot(1,2,2)
sns.boxplot(df.cluster, df.severity)
```
